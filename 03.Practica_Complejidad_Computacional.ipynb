### Ejercicio 1: Operaci√≥n constante (O(1))
Tipo de complejidad: O(1)


```python
import time
import matplotlib.pyplot as plt
import numpy as np

def operacion_constante(n):
    return n ** 2 + n + 1

n_values = np.arange(1, 10001, 100)
times = []

for n in n_values:
    start_time = time.time()
    operacion_constante(n)
    end_time = time.time()
    times.append(end_time - start_time)

plt.plot(n_values, times)
plt.title('Complejidad O(1)')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.show()
```

### Ejercicio 2: Suma de los primeros n n√∫meros (O(n))
Tipo de complejidad: O(n)


```python
import timeit
import matplotlib.pyplot as plt
import numpy as np

def suma_numeros(n):
    suma = 0
    for i in range(1, n+1):
        suma += i
    return suma

n_values = np.arange(1000, 100001, 2000)
times = []

for n in n_values:
    tiempo = timeit.timeit(lambda: suma_numeros(n), number=100)
    times.append(tiempo / 100)

plt.plot(n_values, times)
plt.title('Complejidad O(n)')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.show()
```

### Ejercicio 3: Suma de una matriz (O(n¬≤))
Tipo de complejidad: O(n¬≤)


```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def suma_matriz(n):
    matriz = np.ones((n, n))
    suma = 0
    for i in range(n):
        for j in range(n):
            suma += matriz[i][j]
    return suma

n_values = np.arange(10, 501, 20)
times = []

for n in n_values:
    tiempo = timeit.timeit(lambda: suma_matriz(n), number=10)
    times.append(tiempo / 10)

# Generar el gr√°fico
plt.plot(n_values, times)
plt.title('Complejidad O(n¬≤) - Suma de matriz')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.show()
```

### Ejercicio 4: Fibonacci recursivo (O(2‚Åø))
Tipo de complejidad: O(2‚Åø)


```python
import timeit
import matplotlib.pyplot as plt
import numpy as np

def O_2n(n):
    if n <= 1:
        return n
    else:
        return O_2n(n-1) + O_2n(n-2)

n_values = np.arange(1, 31, 2)
times = []

for n in n_values:
    tiempo = timeit.timeit(lambda: O_2n(n), number=1)
    times.append(tiempo)

plt.plot(n_values, times)
plt.title('Complejidad O(2‚Åø) - Fibonacci recursivo')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.show()
```

### Ejercicio 5: Multiplicaci√≥n de matrices (O(n¬≥))
Tipo de complejidad: O(n¬≥)


```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def multiplicar_matrices_manual(n):
    A = np.ones((n, n))
    B = np.ones((n, n))
    C = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C

n_values = np.arange(10, 51, 5)
times = []

for n in n_values:
    tiempo = timeit.timeit(lambda: multiplicar_matrices_manual(n), number=5)
    times.append(tiempo / 5)

plt.plot(n_values, times)
plt.title('Complejidad O(n¬≥) - Multiplicaci√≥n de matrices manual')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.grid(True)
plt.show()
```

### Ejercicio 6: Generaci√≥n de permutaciones de una lista (O(n!))

```python
import timeit
import itertools
import matplotlib.pyplot as plt
import numpy as np

def generar_permutaciones(arr):
    return list(itertools.permutations(arr))

n_values = np.arange(1, 9, 1)
times = []

for n in n_values:
    lista = list(range(n))
    tiempo = timeit.timeit(lambda: generar_permutaciones(lista), number=5)
    times.append(tiempo / 5)

plt.plot(n_values, times, marker='o')
plt.yscale('log')
plt.title('Complejidad O(n!) - Generaci√≥n de permutaciones')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (log segundos)')
plt.grid(True, which="both", ls="--")
plt.show()
```

### Ejercicio 7: Multiplicaci√≥n de tres matrices (O(n¬≥))

```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def multiplicar_tres_matrices_manual(n):
    A = np.ones((n, n))
    B = np.ones((n, n))
    C = np.ones((n, n))
    resultado = np.zeros((n, n))
    
    AB = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            for k in range(n):
                AB[i][j] += A[i][k] * B[k][j]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                resultado[i][j] += AB[i][k] * C[k][j]
    
    return resultado

n_values = np.arange(5, 51, 5)
times = []

for n in n_values:
    tiempo = timeit.timeit(lambda: multiplicar_tres_matrices_manual(n), number=3)
    times.append(tiempo / 3)

plt.plot(n_values, times, marker='o')
plt.title('Complejidad O(n¬≥) - Multiplicaci√≥n de tres matrices manual')
plt.xlabel('Tama√±o del problema (n)')
plt.ylabel('Tiempo de ejecuci√≥n (segundos)')
plt.grid(True)
plt.show()
```

### Ejercicio 8: Crear una clase llamada ComplejidadComputacional, donde cada m√©todo representar√° cada uno de los ejercicios, de forma tal que se pueda ejecutar cada algoritmo, graficar el comportamiento de complejidad para cada uno y encapsular los par√°metros y la l√≥gica de los algoritmos en una estructura organizada.

```python
import timeit
import matplotlib.pyplot as plt
import numpy as np

class ComplejidadComputacional:

    def O_uno(self, array):
        # Devuelve el primer elemento del array (complejidad O(1))
        return array[0]

    def O_n(self, valor):
        # Calcula el factorial de un n√∫mero de manera recursiva (complejidad O(n))
        if valor == 0:
            return 1
        else:
            return valor * self.O_n(valor-1)

    def O_n2(self, array):
        # Ordena un array utilizando el algoritmo de burbuja (complejidad O(n^2))
        n = len(array)
        for i in range(n):
            for j in range(0, n-i-1):
                if array[j] < array[j + 1]:
                    array[j], array[j + 1] = array[j + 1], array[j]
        return array

    def O_2n(self, n):
        # Calcula el n-√©simo n√∫mero de Fibonacci de manera recursiva sin memoizaci√≥n (complejidad O(2^n))
        if n <= 1:
            return n
        else:
            return self.O_2n(n-1) + self.O_2n(n-2)

    def O_n2_matriz(self, n):
        # Suma todos los elementos de una matriz de tama√±o n x n (complejidad O(n^2))
        matriz = np.ones((n, n))
        suma = 0
        for i in range(n):
            for j in range(n):
                suma += matriz[i][j]
        return suma

    def O_n3(self, n):
        # Multiplica dos matrices de tama√±o n x n (complejidad O(n^3))
        A = np.ones((n, n))
        B = np.ones((n, n))
        C = np.zeros((n, n))

        for i in range(n):
            for j in range(n):
                for k in range(n):
                    C[i][j] += A[i][k] * B[k][j]
        return C

def medir_tiempos():
    cc = ComplejidadComputacional()
    metodos = ['O_uno', 'O_n', 'O_n2', 'O_2n', 'O_n2_matriz', 'O_n3']
    tiempos = {metodo: [] for metodo in metodos}

    tamanos = range(1, 15)
    for n in tamanos:
        array = np.random.randint(0, 100, size=n)

        tiempos['O_uno'].append(timeit.timeit(lambda: cc.O_uno(array), number=100))
        tiempos['O_n'].append(timeit.timeit(lambda: cc.O_n(n), number=100))
        tiempos['O_n2'].append(timeit.timeit(lambda: cc.O_n2(array), number=100))
        tiempos['O_2n'].append(timeit.timeit(lambda: cc.O_2n(n), number=100))
        tiempos['O_n2_matriz'].append(timeit.timeit(lambda: cc.O_n2_matriz(n), number=10))
        tiempos['O_n3'].append(timeit.timeit(lambda: cc.O_n3(n), number=10))

    return tiempos, tamanos

# Medir tiempos de ejecuci√≥n y generar gr√°fico
tiempos, tamanos = medir_tiempos()

for metodo in tiempos:
    plt.plot(tamanos, tiempos[metodo], label=metodo)

plt.xlabel('Tama√±o de entrada')
plt.ylabel('Tiempo de ejecuci√≥n')
plt.legend()
plt.show()
```
