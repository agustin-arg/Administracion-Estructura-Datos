IMPLEMENTACIÓN
 Un árbol está compuesto por nodos.
 Un nodo contiene un elemento o valor y punteros.
 En árboles binarios se requieren 3 punteros: 
• padre
 • izquierda 
• derecha

```python
class Nodo:
 def init (self, e): 
    self._elemento = e 
    self._padre = None 
    self._izquierda = None 
    self._derecha = None
```

Un La clase árbol tiene que tener si o si definido cual es el 
nodo raíz.
 Al momento de iniciar, se pueden agregar más variables, 
como por el ejemplo el tamaño (cantidad de nodos)

```python
class ArbolBinario:
 def init(self): 
    self._raiz = None
    self._tamano = 0
 def len (self):
    return self._tamano
 def vacio(self):
    if self._tamano == 0:
        return True
 def raiz(self):
    return self._raiz
 def es_raiz(self, n):
    if self._raiz == n:
     return True
def es_hoja(self, n):
    if n._izquierda is None and n._derecha is None:
        return True

def padre(self, n):
    if self.es_raiz(n):
        raise ValueError('n es el nodo raíz')
    return n._padre

def izquierda(self, n):
    if n._izquierda is not None:
        return n._izquierda

def derecha(self, n):
    if n._derecha is not None:
        return n._derecha
def profundidad(self, n):
    if self.es_raiz(n):
        return 0
    else:
        return 1 + self.profundidad(n._padre)

def altura(self, n):
    if self.es_hoja(n):
        return 0
    elif n._izquierda is None:
        return 1 + self.altura(n._derecha)
    elif n._derecha is None:
        return 1 + self.altura(n._izquierda)
    elif self.altura(n._izquierda) > self.altura(n._derecha):
        return 1 + self.altura(n._izquierda)
    else:
        return 1 + self.altura(n._derecha)
def agregar(self, e, n=None):
    if self.vacio():
        n = Nodo(e)
        self._raiz = n
        self._tamano += 1
    elif n is None:
        n = self._raiz
    
    if e < n._elemento and n._izquierda is not None:
        self.agregar(e, n._izquierda)
    elif e < n._elemento and n._izquierda is None:
        new = Nodo(e)
        new._padre = n
        n._izquierda = new
        self._tamano += 1
    elif e > n._elemento and n._derecha is not None:
        self.agregar(e, n._derecha)
    elif e > n._elemento and n._derecha is None:
        new = Nodo(e)
        new._padre = n
        n._derecha = new
        self._tamano += 1
```

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None
        self.derecha = None

class ArbolBinario:
    def __init__(self):
        self.raiz = None

    def agregar(self, valor):
        if self.raiz is None:
            self.raiz = Nodo(valor)
        else:
            self._agregar_recursivo(valor, self.raiz)

    def _agregar_recursivo(self, valor, nodo_actual):
        if valor < nodo_actual.valor:
            if nodo_actual.izquierda is None:
                nodo_actual.izquierda = Nodo(valor)
            else:
                self._agregar_recursivo(valor, nodo_actual.izquierda)
        else:
            if nodo_actual.derecha is None:
                nodo_actual.derecha = Nodo(valor)
            else:
                self._agregar_recursivo(valor, nodo_actual.derecha)

    def en_orden(self):
        self._en_orden_recursivo(self.raiz)

    def _en_orden_recursivo(self, nodo_actual):
        if nodo_actual is not None:
            self._en_orden_recursivo(nodo_actual.izquierda)
            print(nodo_actual.valor)
            self._en_orden_recursivo(nodo_actual.derecha)

    def esta_balanceado(self):
        return self._calcular_altura(self.raiz) != -1

    def _calcular_altura(self, nodo):
        if nodo is None:
            return 0
        altura_izquierda = self._calcular_altura(nodo.izquierda)
        altura_derecha = self._calcular_altura(nodo.derecha)
        
        # Si alguno de los subárboles está desbalanceado, propagamos el -1 hacia arriba
        if altura_izquierda == -1 or altura_derecha == -1 or abs(altura_izquierda - altura_derecha) > 1:
            return -1
        
        # Altura del nodo actual
        return max(altura_izquierda, altura_derecha) + 1

# Ejemplo de uso:
arbol = ArbolBinario()
arbol.agregar(5)
arbol.agregar(3)
arbol.agregar(7)
arbol.agregar(2)
arbol.agregar(4)
arbol.agregar(6)
arbol.agregar(8)

print("¿El árbol está balanceado?:", arbol.esta_balanceado())
```
